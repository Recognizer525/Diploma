В статье http://juliejosse.com/wp-content/uploads/2018/07/LectureNotesMissing.html#3 прямо сказано о том, что выведенные формулы для Е- и М- шага не вполне работают, если пропуски есть в нескольких столбцах (т.е. признаках) сразу (последнее предложение из скриншота ниже).

Я разобрался, почему это так. Проблемы возникают на Е-шаге при вычислении статистики $s_{22}$.

Статистика $s_{22}$  вычисляется следующим образом:
Пусть столбец №2 - признак с пропущенными значениями.
$I_{OBS}$ - совокупность индексов, которым не соответствуют строки с пропущенными значениями.
$I_{MIS}$ - совокупность индексов, которым соответствуют строки с пропущенными значениями.

$$
s_{22}=\sum_{i=1}^{len(I_{OBS})} y_{i2}^2 + \sum_{j=1}^{len(I_{MIS})} E[y_{j2}^2|y_{j1}]+\sigma_{2|1}
\\i \in I_{obs}, \: j \in I_{MIS}
$$


Как можно догадаться, в случае есть имеется несколько столбцов с пропусками, условная дисперсия заменится блоком ковариационной матрицы, а условное мат.ожидание так и останется 1x1, соответственно сложить их так, чтобы получилось число нельзя.
