```python


import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
 
# Подготовка данных
data = np.array([[1, 2, np.nan],
                 [4, np.nan, 6],
                 [7, 8, 9]])
 
# Замена нан значений на нули
data[np.isnan(data)] = 0
 
# Масштабирование данных
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)
 
# Применение PCA
pca = PCA(n_components=2)
pca_data = pca.fit_transform(scaled_data)
 
# Заполнение недостающих значений
missing_values = np.where(data == 0)
print('missing_values',missing_values)
filled_data = data.copy()
for idx in range(len(missing_values[0])):
    missing_row = missing_values[0][idx]
    missing_col = missing_values[1][idx]
    component = np.argmax(pca.components_[:, missing_col])
    filled_data[missing_row, missing_col] = (
        pca_data[missing_row, component] * np.sqrt(pca.explained_variance_[component])
        + pca.mean_[missing_col]
)
# Обратное преобразование PCA
reconstructed_data = pca.inverse_transform(pca_data)
 
print("Заполненные данные:")
print(filled_data)
print("\nВосстановленные данные:")
print(reconstructed_data)
```
