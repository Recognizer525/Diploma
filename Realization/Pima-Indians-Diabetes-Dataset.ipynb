```python
Full_D_MCAR = {}
Full_D_MCAR['mean_fill'] = mvt.mean_fill(pima1)
#Full_D_MCAR['lr_fill'] = mvt.lr_fill(D_MCAR)
Full_D_MCAR['stoch_lr_fill'] = mvt.lr_fill(pima1, noise=True)
Full_D_MCAR['em_fill'] = mvt.EM(pima1)
Full_D_MCAR['bootstrap_fill'] = mvt.bootstrap_fill(pima1)
Full_D_MCAR['KNN_fill'] = mvt.knn_fill(pima1)
#Full_D_MCAR['KNN_fill2'] = mvt.knn_fill(D_MCAR, weights='distance')
Full_D_MCAR['multiple_fill'] = mvt.multiple_fill(pima1)
Full_D_MCAR['missforest_fill'] = mvt.missforest_fill(pima1)
```

```python
res = solver.estimates(Full_D_MCAR, pima[8]).reshape(-1,1)
```

```python
res
```
